```python
from bs4 import BeautifulSoup
```

```python
import pandas as pd
```

```python
def read_university_ranking(html_path):
    with open(html_path, 'r', encoding='UTF8') as f:
        contents = f.read()
        soup = BeautifulSoup(contents)
        a_list = soup.find_all("a", {"class": "Anchor-byh49a-0 DetailCardGlobalUniversities__StyledAnchor-sc-1v60hm5-5 kQpddJ cTSURq"})
        subject_scores = soup.find_all("dd", {"class": "QuickStatHug__Description-hb1bl8-1 bBQBxy"})
        subject_scores = [float(subject_scores[i].get_text()) for i in range(0, len(subject_scores), 3)]
        univs = [a.get_text() for a in a_list]
        ranks = list(range(1,len(univs)+1))
    
    return univs, subject_scores, ranks
```

```python
univs_cs, subject_scores_cs, ranks_cs = read_university_ranking('cs-2021-top120.html')
univs_neuro, subject_scores_neuro, ranks_neuro = read_university_ranking('neuro-2021-top120.html')
```

```python
univs = list(set(univs_cs).intersection(set(univs_neuro)))
```

```python
data = {
    'univ':list(),
    'subject_score_sum':list(),
    'subject_score_cs':list(),
    'subject_score_neuro':list(),
    'rank_cs':list(),
    'rank_neuro':list()
}
```

```python
univ_indexes_cs = [univs_cs.index(univ) for univ in univs]
univ_indexes_neuro = [univs_neuro.index(univ) for univ in univs]
```

```python
for univ, i_cs, i_neuro in zip(univs, univ_indexes_cs, univ_indexes_neuro):
    data['univ'].append(univ)
    data['subject_score_sum'].append(subject_scores_cs[i_cs] + subject_scores_neuro[i_neuro])
    data['subject_score_cs'].append(subject_scores_cs[i_cs])
    data['subject_score_neuro'].append(subject_scores_neuro[i_neuro])
    data['rank_cs'].append(ranks_cs[i_cs])
    data['rank_neuro'].append(ranks_neuro[i_neuro])   
```

```python
df = pd.DataFrame.from_dict(data)
```

```python
df = df.sort_values(by=['subject_score_sum'], ascending=False, ignore_index=True)
df['rank'] = list(range(1,len(df)+1))
cols = df.columns.tolist()
cols = cols[-1:] + cols[:-1]
df = df[cols] 
```
